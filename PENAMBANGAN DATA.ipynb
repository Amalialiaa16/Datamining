### Menentukan Outlier

Outlier adalah data yang berbeda secara signifikan dari data lain dalam sebuah set data. Mereka bisa menjadi nilai yang sangat kecil atau sangat besar dibandingkan dengan nilai rata-rata atau mayoritas data lainnya.

LOF (Local Outlier Factor) adalah metode statistik yang digunakan untuk mendeteksi outlier dalam sebuah set data. Metode ini menghitung seberapa jauh suatu titik data berada dari tetangga-tetangganya dalam konteks data tersebut. LOF memberikan nilai yang menggambarkan seberapa "aneh" atau tidak biasa suatu titik data dibandingkan dengan tetangganya. 

Berikut merupakan implementasi dari perhitungan LOF dengan menggunakan library Scikit Learn.

```
import pandas as pd
from sklearn.neighbors import LocalOutlierFactor

# Membaca dataset dari file CSV
X = pd.read_csv("UCI_Real_Estate_Valuation.csv")

# Membuat objek LocalOutlierFactor
lof = LocalOutlierFactor(n_neighbors=5, contamination=0.1)

# Prediksi status outlier
y_pred = lof.fit_predict(X)

# Mencetak tabel prediksi status outlier
print("Predicted table:", y_pred)

# Mencetak nilai LOF yang bersifat negatif
print("Negative LOF scores:", -lof.negative_outlier_factor_)

# Cari indeks dari nilai -1 dalam array y_pred
outlier_indices = [index for index, value in enumerate(y_pred) if value == -1]

# Cetak indeks dari nilai -1 untuk mengetahui data ke berapa yang dianggap sebagai outlier
print("Data outlier terdapat pada indeks:", outlier_indices)

# Kesimpulan
print("KESIMPULAN: Disini asumsinya adalah 0.1 (10%) dari seluruh data dianggap outlier. Output -1 adalah ciri-ciri dari outlier. Dan output yang menunjukkan angka-angka desimal yang lebih tinggi dari angka-angka desimal lainnya merupakan outliernya. Hal ini ditunjukkan pada indeks ke", outlier_indices)
```

KESIMPULAN: Disini asumsinya adalah 0.1(10%) dari seluruh data dianggap outlier. output -1 adalah ciri-ciri dari outlier. dan output yang menunjukkan angka-angka desimal yang lebih tinggi dari angka-angka desimal lainnya merupakan outliernya. hal ini ditunjukkan pada indeks ke [0, 6, 8, 9, 11, 13, 14, 19, 20, 25, 29, 31, 34, 51, 57, 66, 89, 92, 116, 117, 118, 130, 141, 145, 148, 156, 168, 169, 194, 249, 255, 270, 275, 283, 298, 318, 347, 352, 354, 359, 391, 407]

### Hasil Hitung presentase outlier

```
# Hitung jumlah total sampel
total_samples = len(X)

# Hitung jumlah sampel yang dianggap sebagai outlier
outlier_samples = sum(1 for value in y_pred if value == -1)

# Hitung persentase sampel yang dianggap outlier
outlier_percentage = (outlier_samples / total_samples) * 100

# Cetak hasil perhitungan
print(f"Total samples: {total_samples}")
print(f"Outlier samples: {outlier_samples}")
print(f"Outlier percentage: {outlier_percentage:.2f}%")
```

### Menghapus Outlier

Keberadaan outlier dapat mempengaruhi hasil analisis data, sehingga perlu untuk dilakukan  deteksi dan penghapusan outlier. Berikut merupakan program untuk menghapus outlier:

```
# Mengimpor pustaka pandas dengan alias 'pd'.
import pandas as pd

# Mengimpor kelas LocalOutlierFactor dari modul neighbors di pustaka Scikit-learn (sklearn).
from sklearn.neighbors import LocalOutlierFactor

# Membaca dataset dari file CSV ("dataset.csv") ke dalam DataFrame X menggunakan Pandas.
X = pd.read_csv("UCI_Real_Estate_Valuation.csv")

# Membuat objek LocalOutlierFactor (LOF).
# n_neighbors=5 menentukan jumlah tetangga yang akan digunakan dalam perhitungan LOF.
# contamination=0.1 menentukan tingkat kontaminasi atau persentase outlier yang diharapkan dalam data.
lof = LocalOutlierFactor(n_neighbors=5, contamination=0.1)

# Menggunakan metode fit_predict() dari objek LOF untuk menentukan status outlier (outlier atau bukan) untuk setiap sampel dalam data.
# Hasilnya akan berupa array yang berisi prediksi status outlier untuk setiap sampel.
y_pred = lof.fit_predict(X)

# Cari indeks dari nilai -1 dalam array y_pred
outlier_indices = [index for index, value in enumerate(y_pred) if value == -1]

# Cetak indeks dari nilai -1 untuk mengetahui data ke berapa yang dianggap sebagai outlier
print("Data outlier terdapat pada indeks:", outlier_indices)

# Menghapus baris yang mengandung outlier dari DataFrame
X_cleaned = X.drop(outlier_indices)

# Menyimpan DataFrame yang telah dibersihkan ke file CSV baru
X_cleaned.to_csv("real estate_tanpa_outlier.csv", index=False)

# Menampilkan jumlah baris asli dan jumlah baris setelah outlier dihapus
print("Jumlah baris asli:", len(X))
print("Jumlah baris setelah outlier dihapus:", len(X_cleaned))
print("Dataset tanpa outlier telah disimpan ke 'real estate_tanpa_outlier.csv'")
```

Kesimpulan : Kode ini membaca dataset, mendeteksi outlier menggunakan metode Local Outlier Factor, lalu menghapus baris yang mengandung outlier, dan menyimpan dataset yang telah dibersihkan ke file baru yang bernama real estate_tanpa_outlier.csv. Jumlah baris asli adalah 414 baris, setelah outlier dihapus menjadi 372 baris.

### Menentukan nilai K

Program ini digunakan untuk melakukan evaluasi dan membandingkan kinerja model K-Nearest Neighbors (K-NN) dengan jumlah tetangga (k) yang berbeda  untuk memilih nilai k yang optimal untuk model. Kami menggunakan teknik Cross Validation. Cross-Validation adalah metode yang digunakan untuk mengevaluasi kinerja model dengan cara membagi data menjadi subset training dan subset testing secara berulang-ulang.

```
import pandas as pd
import pickle
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import MinMaxScaler, KBinsDiscretizer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Membaca data dari file CSV
data = pd.read_csv('UCI_Real_Estate_Valuation.csv')

# Memisahkan data menjadi fitur (X) dan target (y)
X = data.drop('Y house price of unit area', axis=1)
y = data['Y house price of unit area']

# Mengubah target menjadi kategori berdasarkan rentang harga
bins = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='quantile')
y_bins = bins.fit_transform(y.values.reshape(-1, 1)).astype(int)

# Normalisasi data menggunakan MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Menyimpan scaler ke file pickle
scaler_filename = "classification_preprocessing.pickle"
pickle.dump(scaler, open(scaler_filename, "wb"))

# Membagi data menjadi data pelatihan dan data pengujian
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_bins, test_size=0.2, random_state=42)

# Menentukan kisaran nilai K yang ingin diuji
param_grid = {'n_neighbors': [1, 3, 5, 7, 9, 11, 13, 15]}

# Inisialisasi model KNN
knn = KNeighborsClassifier()

# Inisialisasi Grid Search Cross-Validation
grid_search = GridSearchCV(estimator=knn, param_grid=param_grid, cv=5)

# Melatih model menggunakan Grid Search Cross-Validation
grid_search.fit(X_train, y_train)

# Menampilkan hasil Grid Search
print("K optimal: ", grid_search.best_params_)

# Menggunakan model terbaik untuk prediksi pada data uji
best_knn = grid_search.best_estimator_
y_pred = best_knn.predict(X_test)

# Menghitung dan menampilkan akurasi model
accuracy = accuracy_score(y_test, y_pred)
print("Akurasi pada data uji: ", accuracy)

# Menyimpan model yang sudah dilatih ke file pickle
model_filename = "knn_classification_model.pickle"
pickle.dump(best_knn, open(model_filename, "wb"))
```

KESIMPULAN : dari program diatas dapat disimpulkan bahwa nilai K yang optimal adalah K=3, dengan akurasi data :0.7228915662650602

### Menghitung KNN

 langkah selanjutnya adalah menghitung akurasi KNN dengan nilai K yang telah dihitung

```
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, KBinsDiscretizer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Membaca data dari file CSV
data = pd.read_csv('UCI_Real_Estate_Valuation.csv')

# Memisahkan data menjadi fitur (X) dan target (y)
X = data.drop('Y house price of unit area', axis=1)  # Jika targetnya disebut 'target'
y = data['Y house price of unit area']

# Mengubah target menjadi kategori menggunakan KBinsDiscretizer
bins = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='quantile')
y_bins = bins.fit_transform(y.values.reshape(-1, 1)).astype(int).ravel()

# Normalisasi data menggunakan MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Membagi data menjadi data pelatihan dan data pengujian
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_bins, test_size=0.2, random_state=42)

# Menginisialisasi model KNN
knn = KNeighborsClassifier(n_neighbors=3)

# Melatih model KNN
knn.fit(X_train, y_train)

# Melakukan prediksi dengan data pengujian
y_pred = knn.predict(X_test)

# Menghitung akurasi model
accuracy = accuracy_score(y_test, y_pred)
print("Akurasi model KNN:", accuracy)
```

KESIMPULAN: Nilai 0.7228915662650602 menunjukkan nilai  akurasi dari model K-NN pada dataset yang digunakan. Akurasi adalah ukuran seberapa baik model dapat memprediksi kelas yang benar dari data uji yang belum pernah dilihat sebelumnya. Semakin tinggi nilai output ini (mendekati 1), semakin baik model K-NN bekerja dalam memprediksi kelas yang benar.

n_estimator selanjutnya dicari dengan prediksi keakuratan tertinggi dan nilai n_estimator terkecil. Berikut merupakan program untuk menampilkan nilai dan grafik n_estimator yang optimal:

```
import pandas as pd
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Load the training and testing datasets
train_data = pd.read_csv('realestate_training_80.csv')
test_data = pd.read_csv('realestate_testing_20.csv')

# Assuming the last column is the target
X_train = train_data.iloc[:, :-1]
y_train = train_data.iloc[:, -1]
X_test = test_data.iloc[:, :-1]
y_test = test_data.iloc[:, -1]

# Extract feature names
feature_names = X_train.columns

# Sample input for prediction with correct feature names
sample_input = pd.DataFrame([[0.0582010582, 0, 0.5, 0.3301886792, 0, 0.09677419355, 1, 0]], columns=feature_names)

# Initialize lists to store the number of estimators and corresponding errors
estimators = list(range(3, 20))
mean_squared_errors = []

# Loop through n_estimators from 3 to 20
for n_estimators in estimators:
    # Create and train the BaggingRegressor with a GaussianProcessRegressor estimator
    clf = BaggingRegressor(base_estimator=GaussianProcessRegressor(),
                           n_estimators=n_estimators,
                           random_state=0).fit(X_train, y_train)
    
    # Make a prediction
    sample_input_aligned = sample_input[X_train.columns]
    prediction = clf.predict(sample_input_aligned)
    print(f'Prediction with n_estimators={n_estimators}: {prediction}')
    
    # Evaluate the model on the testing set
    y_pred = clf.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mean_squared_errors.append(mse)
    print(f'Mean Squared Error with n_estimators={n_estimators}: {mse:.4f}')

# Plot the mean squared errors
plt.figure(figsize=(10, 6))
plt.plot(estimators, mean_squared_errors, marker='o')
plt.xlabel('Number of Estimators')
plt.ylabel('Mean Squared Error')
plt.title('Bagging Regressor vs. Number of Estimators')
plt.ylim(0, max(mean_squared_errors) + 1)
plt.grid(True)
plt.show()
```

```
import pandas as pd
import pickle
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Membaca data dari file CSV
data = pd.read_csv('UCI_Real_Estate_Valuation.csv')

# Memisahkan data menjadi fitur (X) dan target (y)
X = data.drop('Y house price of unit area', axis=1)  # Jika targetnya disebut 'target'
y = data['Y house price of unit area']

# Normalisasi data menggunakan MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Menyimpan scaler ke file pickle
scaler_filename = "preprocessing_realestate.pickle"
pickle.dump(scaler, open(scaler_filename, "wb"))

# Membagi data menjadi data pelatihan dan data pengujian
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Menentukan kisaran nilai K yang ingin diuji
param_grid = {'n_neighbors': [1, 3, 5, 7, 9, 11, 13, 15]}

# Inisialisasi model KNN untuk regresi
knn = KNeighborsRegressor()

# Inisialisasi Grid Search Cross-Validation dengan parameter error_score='raise'
grid_search = GridSearchCV(estimator=knn, param_grid=param_grid, cv=5, error_score='raise')

try:
    # Melatih model menggunakan Grid Search Cross-Validation
    grid_search.fit(X_train, y_train)

    # Menampilkan hasil Grid Search
    print("K optimal: ", grid_search.best_params_)

    # Menggunakan model terbaik untuk prediksi pada data uji
    best_knn = grid_search.best_estimator_
    y_pred = best_knn.predict(X_test)


    # Menghitung dan menampilkan akurasi pada data uji
    r2 = r2_score(y_test, y_pred)
    print("Akurasi Data Uji: ", r2)

    # Menyimpan model yang sudah dilatih ke file pickle
    model_filename = "knn_model_realestate.pickle"
    pickle.dump(best_knn, open(model_filename, "wb"))

except ValueError as e:
    print("Error during model training: ", e)
```

```
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler
import numpy as np
import pickle
from sklearn.utils import resample

# Load the training and testing datasets
train_data = pd.read_csv('realestate_training_80.csv')
test_data = pd.read_csv('realestate_testing_20.csv')

# Assuming the last column is the target
X_train = train_data.iloc[:, :-1]
y_train = train_data.iloc[:, -1]
X_test = test_data.iloc[:, :-1]
y_test = test_data.iloc[:, -1]

# Check for NaN values and handle them if present
X_train = X_train.fillna(X_train.mean())
X_test = X_test.fillna(X_test.mean())

# Convert continuous targets to categorical (e.g., by binning)
y_train_binned = pd.cut(y_train, bins=5, labels=False)
y_test_binned = pd.cut(y_test, bins=5, labels=False)

# Check class balance
print("Class distribution in training set:")
print(y_train_binned.value_counts())

print("Class distribution in test set:")
print(y_test_binned.value_counts())

# Ensure there are no classes with zero samples after binning
if (y_train_binned.value_counts() == 0).any() or (y_test_binned.value_counts() == 0).any():
    raise ValueError("One of the classes has zero samples after binning. Adjust the binning strategy.")

# Resampling to handle class imbalance
y_train_binned = y_train_binned.astype(str)  # Convert to string for resampling
train_data_resampled = pd.concat([X_train, y_train_binned], axis=1)

# Upsample minority classes
majority_class = train_data_resampled[train_data_resampled[y_train_binned.name] == train_data_resampled[y_train_binned.name].mode()[0]]
minority_classes = train_data_resampled[train_data_resampled[y_train_binned.name] != train_data_resampled[y_train_binned.name].mode()[0]]

upsampled_minority_classes = []
for label in minority_classes[y_train_binned.name].unique():
    minority_class = minority_classes[minority_classes[y_train_binned.name] == label]
    upsampled_minority_class = resample(minority_class,
                                        replace=True,  # sample with replacement
                                        n_samples=len(majority_class),  # to match majority class
                                        random_state=123)
    upsampled_minority_classes.append(upsampled_minority_class)

train_data_balanced = pd.concat([majority_class] + upsampled_minority_classes)

X_train_balanced = train_data_balanced.drop(columns=[y_train_binned.name])
y_train_balanced = train_data_balanced[y_train_binned.name].astype(int)  # Convert back to integer

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_balanced)
X_test_scaled = scaler.transform(X_test)

# Ensure there are no features with zero variance
if (np.var(X_train_scaled, axis=0) == 0).any():
    raise ValueError("One of the features has zero variance after scaling. Check your data.")

# Check for any zero values in scaled data
print("Zero values in X_train_scaled:", np.sum(X_train_scaled == 0))
print("Zero values in X_test_scaled:", np.sum(X_test_scaled == 0))

# Save the scaler
scaler_filename = "preprocessing_realestate.pickle"
with open(scaler_filename, 'wb') as scaler_file:
    pickle.dump(scaler, scaler_file)

# Number of estimators
n_estimators = 10

# Create and train the BaggingClassifier with a Naive Bayes estimator
clf = BaggingClassifier(base_estimator=GaussianNB(), n_estimators=n_estimators, random_state=0)

try:
    clf.fit(X_train_scaled, y_train_balanced)
except ValueError as e:
    print("Error during model training: ", e)
    raise

# Save the trained model to a pickle file
model_filename = "bagging_classifier_model_realestate.pickle"
with open(model_filename, 'wb') as model_file:
    pickle.dump(clf, model_file)

# Make a prediction on the sample input from the training data
sample_input = X_train.iloc[0].values.reshape(1, -1)
sample_input_scaled = scaler.transform(sample_input)
prediction = clf.predict(sample_input_scaled)
print(f'Prediction with n_estimators={n_estimators} for sample input: {prediction}')

# Evaluate the model on the testing set
y_pred = clf.predict(X_test_scaled)
accuracy = accuracy_score(y_test_binned, y_pred)
print(f'Accuracy with n_estimators={n_estimators}: {accuracy:.4f}')

# Classification report for detailed analysis
print("Classification Report:")
print(classification_report(y_test_binned, y_pred))

# Save the results to a DataFrame and then to a CSV file
results_df = pd.DataFrame({
    'n_estimators': [n_estimators],
    'accuracy': [accuracy]
})
results_df.to_csv('bagging_realestate.csv', index=False)

# Generate random data for predictions
np.random.seed(0)  # Set seed for reproducibility
random_data = np.random.rand(len(X_train), X_train.shape[1])
random_data_scaled = scaler.transform(random_data)

# Make predictions on the random data
random_predictions = clf.predict(random_data_scaled)

# Save the final model's predictions on the entire training set along with Bagging value
train_predictions_df = pd.DataFrame({
    'bagging': [n_estimators] * len(random_predictions),
    'prediction': random_predictions
})
train_predictions_df.to_csv('train_predictions.csv', index=False)
```
